````python
# Queue class: can be implemented using a list in Python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)
    
def Dec_to_Bin_Rec(n):
    ans = ""
    if n >= 1:       
        ans =  Dec_to_Bin_Rec(n // 2)+str(n % 2) 
        return str(ans)
    else :
        return ""

def generateBinaryNumbers(n):
    q = Queue() 
    i=1
    while (i<=n):
        q.enqueue(Dec_to_Bin_Rec(i))
        i+=1
    return q.items[::-1]

n=10
print("When n is,",n,", the binary numbers are:")
print(generateBinaryNumbers(n))

n=13
print("When n is,",n,", the binary numbers are:")
print(generateBinaryNumbers(n))
````
